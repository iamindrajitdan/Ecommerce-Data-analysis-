```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = mysql.connector.connect(
user='root',
password='12345',
host='localhost',
database='MYCUSTOMERES_DB'
)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('SELECT * FROM customers')
customer_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
customer_data
```

```python
cursor.execute('SELECT * FROM product')
product_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
product_data
```

```python
cursor.execute('SELECT * FROM order_details')
order_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in
cursor.description])
order_data
```

```python
print("\ncustomer")
print(customer_data.isnull().sum())
print("\nproduct")
print(product_data.isnull().sum())
print("\norder")
print(order_data.isnull().sum())


```

```python
print("\ncustomer")
print(customer_data.dtypes)
print("\nproduct")
print(product_data.dtypes)
print("\norder")
print(order_data.dtypes)
```

```python
customer_city_count = customer_data.groupby('city')['customer_id'].nunique()
customer_city_count.plot(kind='bar', title='Total Customers City-wise')
plt.ylabel('Number of Customers')
plt.show()
```

```python

frequent_customers = order_data.groupby('customer_id')['order_id'].count().sort_values(ascending=False)
frequent_customers.head(10).plot(kind='bar', title='Top 10 Frequent Customers')
plt.ylabel('Number of Orders')
plt.show()
```

```python

product_category_count = product_data.groupby('category')['product_id'].count()
product_category_count.plot(kind='bar', title='Total Products by Category')
plt.ylabel('Number of Products')
plt.show()
```

```python
low_stock_products = product_data[product_data['stock'] < 10] 
print(low_stock_products[['product_id', 'stock']])
```

```python
top_10_orders = order_data.groupby('product_id')['quantity'].sum().sort_values(ascending=False).head(10)
top_10_orders.plot(kind='bar', title='Top 10 Products by Order Quantity')
plt.ylabel('Total Quantity')
plt.show()
```

```python
order_status_distribution = order_data['order_status'].value_counts()
order_status_distribution.plot(kind='pie', autopct='%1.1f%%', title='Order Status Distribution')
plt.show()
```

```python
total = order_data['total_price'].sum()
print(f"Total revenue from all orders: {total}")
```

```python
highest_ordered_product = order_data.groupby('product_id')['quantity'].sum().idxmax()
lowest_ordered_product = order_data.groupby('product_id')['quantity'].sum().idxmin()

print(f"Highest ordered product: {highest_ordered_product}")
print(f"Lowest ordered product: {lowest_ordered_product}")
```

```python
payment_modes = order_data['payment_mode'].value_counts()
payment_modes.plot(kind='bar', title='Most Preferred Payment Modes')
plt.ylabel('Frequency')
plt.show()
```

```python
city_distribution = customer_data['city'].value_counts()
city_distribution.plot(kind='bar', title='Customer Distribution Across Cities')
plt.ylabel('Number of Customers')
plt.show()
```

```python
best_selling_products = order_data.groupby('product_id')['quantity'].sum().sort_values(ascending=False).head(10)
best_selling_products.plot(kind='bar', title='Top 10 Best Selling Products')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
repeat_customers = order_data.groupby('customer_id')['order_id'].count()
repeat_customers = repeat_customers[repeat_customers > 1]
repeat_customers.plot(kind='bar', title='Repeat Customers')
plt.ylabel('Number of Orders')
plt.show()
```

```python
payment_status_counts = order_data['payment_mode'].value_counts()
payment_status_counts.plot(kind='bar', title='Successful vs Pending Payments')
plt.ylabel('Order Count')
plt.show()
```
